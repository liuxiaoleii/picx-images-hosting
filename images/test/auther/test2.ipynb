```python
# 导入常用库
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("库导入成功！")
```

```python
# 创建示例数据
data = {
    'A': np.random.randn(100),
    'B': np.random.randn(100) + 2,
    'C': np.random.randn(100) - 1
}
df = pd.DataFrame(data)
df.head(10)
```

```python
# 绘制简单图表
plt.figure(figsize=(10, 6))
df.plot(kind='line')
plt.title('示例数据可视化')
plt.xlabel('索引')
plt.ylabel('数值')
plt.grid(True)
plt.show()
```
